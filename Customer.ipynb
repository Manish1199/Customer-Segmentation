```python
# select features for the clustering

x=df[['Age','Annual Income (k$)','Spending Score (1-100)']]
```

```python
scaller=StandardScaler()
```

```python
x_scalled=scaller.fit_transform(x)
```

```python
kmeans=KMeans(n_clusters=5,random_state=0)
```

```python
df['clusters']=kmeans.fit_predict(x_scalled)
```

```python
df.head()
```

```python
# visualize clusters

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='clusters', data=df, palette='Set1', s=100)
plt.title('Customer Segmentation')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
score = silhouette_score(x_scalled, df['clusters'])
print(f'Silhouette Score: {score}')
```

```python
dictionaries=[]
K = range(1,100)
for k in K:
    kmeans=KMeans(n_clusters=k)
    kmeans.fit_transform(x_scalled)
    dictionaries.append(kmeans.inertia_)
plt.figure(figsize=(8, 6))
plt.plot(K, dictionaries, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# profile cluster
cluster_profile=df.groupby('clusters').mean(numeric_only=True)
cluster_profile
```

```python
# Calculating mode for 'Gender' within each cluster
gender_mode = df.groupby('clusters')['Gender'].agg(lambda x: x.mode()[0])
print(gender_mode)
```
